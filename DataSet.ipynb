```python
#สัดส่วนผู้รอดชีวิตผู้ชายกับผู้หญฺิง
import numpy as np
print(df['sex'].value_counts())

print('% of male who survived',100*np.mean(df['survived'][df['sex']=='male']))
print('% of female who survived',100*np.mean(df['survived'][df['sex']=='female']))
```

```python
#percent of pclass [1:firstclass,2:secondclass,3:thirdclass]
#who survived
print('% of pclass who survived',100*np.mean(df['survived'][df['pclass']==1]))
print('% of pclass who survived',100*np.mean(df['survived'][df['pclass']==2]))
print('% of pclass who survived',100*np.mean(df['survived'][df['pclass']==3]))
```

```python
#ค่าเฉลี่ยผู้รอดชีวิต
print(100*df[['pclass','survived']].groupby(['pclass']).mean())
```

```python
#copy Dataframe
df2=df.copy()
#numbers of missing row for each column
df2.isnull().sum()
```

```python
#replace mean for missing in age column
df2['age'].fillna(df2['age'].mean(),inplace=True)
df2.isnull().sum()
```

```python
#drop missing in embark column
df2.dropna(subset=['embarked'],inplace=True)
df2.isnull().sum()
```

```python
print(df2["cabin"].value_counts())
```

```python
#replace missing whit mode for cabin column
df2['cabin'].fillna(df2['cabin'].mode()[0],inplace=True)
df2.isnull().sum()
```

```python
df2.dropna(subset=['fare'],inplace=True)
df2.isnull().sum()
```

```python
df2.drop(['boat','body','home.dest'],axis=1,inplace=True)
df2.isnull().sum()
```
